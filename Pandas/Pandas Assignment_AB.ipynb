### Q9. How do you concatenate two Pandas DataFrames?

```python
# Firstly creating two dataframes
data1 = {'Name':['Sachin', 'Sourav', 'Rahul', 'Laxman'], 
        'Age':[49, 50, 50, 48], 
        'Address':['Mumbai', 'Kolkata', 'Karnataka', 'Hyderabad'], 
        'Total ODI Run':[18426, 11363, 10889, 2338]
        } 
   
data2 = {'Name':['Dhoni', 'Yuvi', 'Sehwag'], 
        'Age':[41, 41, 44], 
        'Address':['Ranchi', 'Punjab', 'Delhi'], 
        'Total ODI Run':[10773, 11363, 8273]
        } 
 
# Convert the dictionary into DataFrame  
df1 = pd.DataFrame(data1,index=[0, 1, 2, 3])
 
df2 = pd.DataFrame(data2, index=[4, 5, 6])
     
```

```python
df1
```

```python
df2
```

```python
# using a .concat() method
frames = [df1, df2] 
df3 = pd.concat(frames)
print(df3)
```

### Q10. How do you merge two Pandas DataFrames on a specific column?

```python
df4=pd.merge(df1,df2,on="Total ODI Run")
df4
```

### Q25. What is the difference between .loc and .iloc in Pandas?

ANS: 1) loc is Label-based data selector where iloc is Index-based data selector.  
     2) loc Indices should be sorted in order, or loc[ ] will only select the mentioned indices when slicing where         iloc Indices need not be sorted in order when slicing.  
     3) loc Indices should be numerical, else slicing cannot be done where iloc Indices can be numerical or     categorical.  
     4) In loc the end index is included during slicing where in iloc the end index is excluded during slicing.  
     5) loc accepts bool series or list in conditions where iloc Only accepts bool list in conditions.
